```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.preprocessing import MinMaxScaler
import numpy as np
from statsmodels.tsa.stattools import adfuller

# Load dataset
file_path = r"C:\Users\vaish\Project phase I\Supermart Grocery Sales - Retail Analytics Dataset.csv"
data = pd.read_csv(file_path)

# Convert 'Order Date' to datetime
data['Order Date'] = pd.to_datetime(data['Order Date'], errors='coerce')

# Handle missing and non-numeric sales
data['Sales'] = pd.to_numeric(data['Sales'], errors='coerce')
data = data.dropna(subset=['Order Date', 'Sales'])

# Aggregate sales by date and resample to monthly frequency
time_series_data = data.groupby('Order Date')['Sales'].sum()
time_series_data = time_series_data.resample('ME').sum()  # 'ME' for month-end frequency
time_series_data = time_series_data.fillna(0)  # Fill missing values

# Check stationarity and apply differencing if needed
result = adfuller(time_series_data)
if result[1] > 0.05:  # Non-stationary
    time_series_data = time_series_data.diff().dropna()

# Define and fit SARIMA model (simplify if data is sparse)
sarima_model = SARIMAX(time_series_data, order=(1, 1, 1), seasonal_order=(0, 1, 0, 12))
sarima_fit = sarima_model.fit(disp=False)

# Forecast future sales (e.g., next 12 months)
sarima_pred = sarima_fit.forecast(steps=12)

# Create forecast index
forecast_index = pd.date_range(start=time_series_data.index[-1] + pd.Timedelta(days=1), periods=12, freq='ME')

# Plotting actual vs forecasted sales
plt.figure(figsize=(10, 6))
plt.plot(time_series_data, label='Actual Sales')
plt.plot(forecast_index, sarima_pred, label='Forecasted Sales', color='red')

plt.title('Actual vs Forecasted Sales')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.tsa.stattools import adfuller

# Load dataset
file_path = r"C:\Users\vaish\Project phase I\Supermart Grocery Sales - Retail Analytics Dataset.csv"
data = pd.read_csv(file_path)

# Convert 'Order Date' to datetime
data['Order Date'] = pd.to_datetime(data['Order Date'], errors='coerce')

# Handle missing and non-numeric sales
data['Sales'] = pd.to_numeric(data['Sales'], errors='coerce')
data = data.dropna(subset=['Order Date', 'Sales'])

# Aggregate sales by date and resample to monthly frequency
time_series_data = data.groupby('Order Date')['Sales'].sum()
time_series_data = time_series_data.resample('ME').sum()  # 'ME' for month-end frequency
time_series_data = time_series_data.fillna(0)  # Fill missing values

# Check stationarity and apply differencing if needed
result = adfuller(time_series_data)
if result[1] > 0.05:  # Non-stationary
    time_series_data = time_series_data.diff().dropna()

# Define and fit SARIMA model
sarima_model = SARIMAX(time_series_data, order=(1, 1, 1), seasonal_order=(0, 1, 0, 12))
sarima_fit = sarima_model.fit(disp=False)

# Forecast next 3 months
sarima_pred = sarima_fit.forecast(steps=3)

# Create forecast index for next 3 months
forecast_index = pd.date_range(start=time_series_data.index[-1] + pd.Timedelta(days=1), periods=3, freq='ME')

# Plot actual vs 3-month forecasted sales
plt.figure(figsize=(10, 6))
plt.plot(time_series_data, label='Actual Sales', marker='o')
plt.plot(forecast_index, sarima_pred, label='3-Month Forecasted Sales', color='red', marker='o')

plt.title('Actual vs 3-Month Forecasted Sales')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.grid()
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from tensorflow.keras.optimizers import Adam
import matplotlib.pyplot as plt

# Load dataset
file_path = r"C:\Users\vaish\Project phase I\Supermart Grocery Sales - Retail Analytics Dataset.csv"
data = pd.read_csv(file_path)

# Convert 'Order Date' to datetime if it's not already
data['Order Date'] = pd.to_datetime(data['Order Date'])

# Aggregate sales by date and resample to monthly frequency
time_series_data = data.groupby('Order Date')['Sales'].sum()
time_series_data = time_series_data.resample('M').sum()  # Resample by month

# Fill missing values if any
time_series_data = time_series_data.fillna(0)

# Scale the data using MinMaxScaler for LSTM input
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(time_series_data.values.reshape(-1, 1))

# Create sequences for LSTM (using 12 months of data to predict the next month)
def create_sequences(data, sequence_length=12):
    X, y = [], []
    for i in range(len(data) - sequence_length):
        X.append(data[i:i + sequence_length])
        y.append(data[i + sequence_length])
    return np.array(X), np.array(y)

sequence_length = 12  # Using 12 months to predict the next month
X, y = create_sequences(scaled_data, sequence_length)

# Reshape X to fit LSTM input shape: (samples, timesteps, features)
X = X.reshape(X.shape[0], X.shape[1], 1)

# Split into train and test sets (80% train, 20% test)
train_size = int(len(X) * 0.8)
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]

# Define the LSTM model
model = Sequential()
model.add(LSTM(units=64, activation='relu', input_shape=(X_train.shape[1], X_train.shape[2])))
model.add(Dense(1))

# Compile the model
model.compile(optimizer=Adam(), loss='mean_squared_error')

# Train the model with multiple epochs
epochs = 50  # Number of epochs
history = model.fit(X_train, y_train, epochs=epochs, batch_size=16, validation_data=(X_test, y_test))

# Predict using the trained LSTM model
predicted_sales = model.predict(X_test)

# Inverse transform the predictions and actual values to get the original scale
predicted_sales_original = scaler.inverse_transform(predicted_sales)
y_test_original = scaler.inverse_transform(y_test.reshape(-1, 1))

# Plot Actual vs Predicted Sales
plt.figure(figsize=(10, 6))
plt.plot(y_test_original, label='Actual Sales', color='blue')
plt.plot(predicted_sales_original, label='Predicted Sales', color='orange')
plt.title('Actual vs Predicted Sales')
plt.xlabel('Time')
plt.ylabel('Sales')
plt.legend()
plt.show()

# Print the performance metrics
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

rmse = np.sqrt(mean_squared_error(y_test_original, predicted_sales_original))
mae = mean_absolute_error(y_test_original, predicted_sales_original)
r2 = r2_score(y_test_original, predicted_sales_original)

print(f"RMSE: {rmse}")
print(f"MAE: {mae}")
print(f"R2 Score: {r2}")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, Input
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Load dataset
file_path = r"C:\Users\vaish\Project phase I\Supermart Grocery Sales - Retail Analytics Dataset.csv"
data = pd.read_csv(file_path)

# Convert 'Order Date' to datetime and aggregate sales by month
data['Order Date'] = pd.to_datetime(data['Order Date'], errors='coerce')
data.dropna(subset=['Order Date'], inplace=True)
data['Sales'] = pd.to_numeric(data['Sales'], errors='coerce')

# Aggregate and resample sales data to monthly frequency
monthly_sales = data.set_index('Order Date').resample('ME')['Sales'].sum().fillna(0)

# Scale data
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(monthly_sales.values.reshape(-1, 1))

# Prepare dataset for LSTM
def create_dataset(data, time_step=1):
    X, y = [], []
    for i in range(len(data) - time_step - 1):
        X.append(data[i:(i + time_step), 0])
        y.append(data[i + time_step, 0])
    return np.array(X), np.array(y)

time_step = 12
X, y = create_dataset(scaled_data, time_step)
X = X.reshape(X.shape[0], X.shape[1], 1)

# Split data into training and testing
train_size = int(len(X) * 0.8)
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]

# Build LSTM model with Input layer
model = Sequential([
    Input(shape=(time_step, 1)),
    LSTM(50, return_sequences=True),
    LSTM(50, return_sequences=False),
    Dense(25),
    Dense(1)
])
model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
history = model.fit(X_train, y_train, batch_size=32, epochs=50, verbose=1)

# Predict
train_predict = model.predict(X_train)
test_predict = model.predict(X_test)

# Transform predictions back to original scale
train_predict = scaler.inverse_transform(train_predict)
test_predict = scaler.inverse_transform(test_predict)
y_train = scaler.inverse_transform(y_train.reshape(-1, 1))
y_test = scaler.inverse_transform(y_test.reshape(-1, 1))

# Forecast future sales for 12 months
forecast_input = scaled_data[-time_step:].reshape(1, time_step, 1)
forecast = []
for _ in range(12):
    predicted_value = model.predict(forecast_input, verbose=0)
    forecast.append(predicted_value[0, 0])
    forecast_input = np.append(forecast_input[:, 1:, :], [[predicted_value]], axis=1)

forecast = scaler.inverse_transform(np.array(forecast).reshape(-1, 1))
forecast_index = pd.date_range(start='2024-01-01', periods=12, freq='ME')

# Plot original data, predictions, and forecast
plt.figure(figsize=(12, 6))
plt.plot(monthly_sales, label='Actual Sales', color='blue')
plt.plot(monthly_sales.index[:len(train_predict)], train_predict, label='Train Prediction', color='green')
plt.plot(monthly_sales.index[len(train_predict):(len(train_predict) + len(test_predict))], test_predict, label='Test Prediction', color='red')
plt.plot(forecast_index, forecast, label='Forecast (2024)', color='orange')
plt.title("LSTM Sales Prediction")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.legend()
plt.show()

# Evaluation Metrics
rmse = np.sqrt(mean_squared_error(y_test, test_predict))
mae = mean_absolute_error(y_test, test_predict)
r2 = r2_score(y_test, test_predict)

print(f"Root Mean Squared Error (RMSE): {rmse}")
print(f"Mean Absolute Error (MAE): {mae}")
print(f"R2 Score: {r2}")
```

```python
import pandas as pd
import numpy as np
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import mean_absolute_percentage_error
from xgboost import XGBRegressor
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import LSTM, Dense
from keras.preprocessing.sequence import TimeseriesGenerator
import warnings

warnings.filterwarnings("ignore")

# Step 1: Load and preprocess data
df = pd.read_csv("Supermart Grocery Sales - Retail Analytics Dataset.csv")
df['Order Date'] = pd.to_datetime(df['Order Date'])
df = df.sort_values('Order Date')
monthly_sales = df.groupby(pd.Grouper(key='Order Date', freq='M'))['Sales'].sum().reset_index()
monthly_sales.columns = ['Date', 'Sales']
monthly_sales = monthly_sales.set_index('Date')

# Step 2: Train-test split
train_size = int(len(monthly_sales) * 0.8)
train, test = monthly_sales[:train_size], monthly_sales[train_size:]

# ---------- SARIMA ----------
sarima_model = SARIMAX(train, order=(1, 1, 1), seasonal_order=(1, 1, 1, 12))
sarima_result = sarima_model.fit(disp=False)
sarima_forecast = sarima_result.forecast(steps=len(test))
sarima_mape = mean_absolute_percentage_error(test, sarima_forecast) * 100

# ---------- XGBoost ----------
xgb_data = monthly_sales.reset_index()
xgb_data['Month'] = xgb_data['Date'].dt.month
xgb_data['Year'] = xgb_data['Date'].dt.year
X = xgb_data[['Year', 'Month']]
y = xgb_data['Sales']
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]

xgb_model = XGBRegressor()
xgb_model.fit(X_train, y_train)
xgb_pred = xgb_model.predict(X_test)
xgb_mape = mean_absolute_percentage_error(y_test, xgb_pred) * 100

# ---------- LSTM ----------
scaler = MinMaxScaler()
scaled_train = scaler.fit_transform(train.values.reshape(-1, 1))
n_input = 3
generator = TimeseriesGenerator(scaled_train, scaled_train, length=n_input, batch_size=1)

lstm_model = Sequential([
    LSTM(50, activation='relu', input_shape=(n_input, 1)),
    Dense(1)
])
lstm_model.compile(optimizer='adam', loss='mse')
lstm_model.fit(generator, epochs=30, verbose=0)

# Forecasting with LSTM
test_scaled = scaler.transform(test.values.reshape(-1, 1))
preds = []
first_eval_batch = scaled_train[-n_input:]
current_batch = first_eval_batch.reshape((1, n_input, 1))

for i in range(len(test)):
    lstm_pred = lstm_model.predict(current_batch, verbose=0)[0]
    preds.append(lstm_pred)
    current_batch = np.append(current_batch[:, 1:, :], [[lstm_pred]], axis=1)

lstm_forecast = scaler.inverse_transform(preds)
lstm_mape = mean_absolute_percentage_error(test, lstm_forecast) * 100

# --------- Output results ---------
print("\nModel Accuracy (MAPE %):")
print(f" SARIMA  : {sarima_mape:.2f}%")
print(f" XGBoost : {xgb_mape:.2f}%")
print(f" LSTM    : {lstm_mape:.2f}%")
```
